```python
#Import various libraries which might be useful for numerical calculations and plotting.

import numpy as np
import matplotlib.pyplot as plt
import math
import random
from tkinter import *
from array import *
import time
import datetime
import os


#Define a function which gives us the energy of a spin.
def spin_energy(spin):

    if spin==-1:
        energy=-1.0
    else:
        energy=1.0
    return energy


# Set up the data we will need.

#Number of attempted moves
number_of_attempts=40
#Define the energy times Boltzmann's constant. 
kT=100.0
#Set up our random number generator with a given seed.
random.seed(151)
#Set up the initial spin.
spin=0

#Make arrays to store the spins and the attempt number.
stored_spins=[]
stored_attempt=[]


#Loop over the specified number of attempts.
#This will start at zero and finish at number_of_attempts-1.
for i in np.arange(0,number_of_attempts):
    #Our trial state is to flip the spin.
    if spin==-1:
            trialspin=1
    else:
            trialspin=-1
    #Calculate energy difference between trial state and our present state.
    energy_change= spin_energy(trialspin)-spin_energy(spin)
    
    #Now invoke Metropolis - this is the meat of the program
    #If the energy change is 0 or negative then accept the trail state 
    #as the new state. 
    if energy_change <= 0:
        spin=trialspin
    else: 
    #Otherwise generate a random number between 0 and 1 and apply Metropolis
        random_number=random.random()
        if random_number <np.exp(-energy_change/kT):
            spin=trialspin
    
    #If the above code fails,  then we keep our spin unchanged.
    
    
    #Now store the spin and the sttempt number in separate arrays.
    stored_spins.append(spin)
    stored_attempt.append(i)
    
        
#Plot the spin versus the number of moves. 
plt.xlabel('Attempts')
plt.ylabel('Spin')
plt.plot(stored_attempt,stored_spins, marker="o", color='navy' )
plt.show()



```
