```python
import numpy as np; import matplotlib.pyplot as plt 
import astropy.constants as c; import astropy.units as u
```

```python
gamma = np.array([2,4,5,7,10])
beta_a = np.sqrt([1-1/gamma**2])
beta_a = np.append(0.5,beta_a)
```

```python
theta = np.linspace(0, np.pi/2, 100)
for beta in beta_a:
    beta_t = beta*np.sin(theta)/(1-beta*np.cos(theta))
    plt.plot(theta,beta_t,label = r'$\beta$ = %.3f'%beta)
plt.plot(theta,[1]*100,'b--')
plt.xlabel(r'$\theta$');plt.ylabel(r'$\beta_t$');plt.legend()
plt.savefig('fig1.pdf')

```
